```python
import pandas as pd

#***** City Level Data Preparation*******

#read citylevel SMB closure data
city_df1=pd.read_csv("Womply Merchants - City - Daily.csv")
city_df1=city_df1[city_df1['cityid'].isin([5,23,25,28])]

#read citylevel SMB revenue data
city_df2=pd.read_csv("Womply Revenue - City - Daily.csv")
city_df2=city_df2[city_df2['cityid'].isin([5,23,25,28])]

#read citylevel spenting revenue data
city_df3=pd.read_csv("Affinity - City - Daily.csv")
city_df3=city_df3[city_df3['cityid'].isin([5,23,25,28])]

#combine all these csvs
city_df4=pd.merge(city_df1,city_df2,on=['year','month','day','cityid'])
city_df=pd.merge(city_df4,city_df3,on=['year','month','day','cityid'])

#add city mapping table
city_map=pd.read_csv("GeoIDs - City.csv")
city_map=city_map[city_map['cityid'].isin([5,23,25,28])]
city_map=city_map[['cityid','cityname']]

#add cityname to city data frame
city_df=pd.merge(city_df,city_map,on='cityid')

# write to csv final results
city_df.to_csv("final_city_level_data.csv")

#***** County Level Data Preparation*******

#read county level SMB closure data
county_df1=pd.read_csv("Womply Merchants - County - Daily.csv")


#read county level SMB revenue data
county_df2=pd.read_csv("Womply Revenue - County - Daily.csv")


#read county level spending data
county_df3=pd.read_csv("Affinity - County - Daily.csv")

#combine all these dataframes
county_df4=pd.merge(county_df1,county_df2,on=['year','month','day','countyfips'])
county_df=pd.merge(county_df4,county_df3,on=['year','month','day','countyfips'],how='left')

#add county mapping table
county_map=pd.read_csv("GeoIDs - County.csv")
county_map=county_map[county_map['stateabbrev'].isin(['AZ','CO','MA','MD'])]
county_map=county_map[['countyfips','countyname','stateabbrev']]
county_df=pd.merge(county_df,county_map,on='countyfips')
county_df


# write to csv final results
county_df.to_csv("final_county_level_data.csv")


```
